```python
import socket
import os

def server(host='localhost', port=12346, output_file='received_server.jpg'):
    buffer_size = 1024
    ack_message = b'ACK'
    timeout = 2  # seconds

    with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as server_socket:
        server_socket.bind((host, port))
        print(f"Listening on {host}:{port}")

        with open(output_file, 'wb') as f:
            expected_seq_num = 0
            server_socket.settimeout(timeout)
            
            while True:
                try:
                    data, addr = server_socket.recvfrom(buffer_size + 4)  # 4 bytes for the sequence number
                    seq_num = int.from_bytes(data[:4], 'big')
                    
                    if seq_num == expected_seq_num:
                        f.write(data[4:])
                        server_socket.sendto(ack_message + seq_num.to_bytes(4, 'big'), addr)
                        expected_seq_num += 1
                        print(f"Received packet {seq_num}, sent ACK")
                    else:
                        # Send ACK for the last correctly received packet
                        server_socket.sendto(ack_message + (expected_seq_num - 1).to_bytes(4, 'big'), addr)
                        print(f"Packet {seq_num} out of order, expected {expected_seq_num}")
                except socket.timeout:
                    print("Waiting for packets...")

if __name__ == "__main__":
    server()
```
